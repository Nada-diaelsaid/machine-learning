import numpy

import theano
import theano.tensor as T

```python
rng= numpy.random
# number of samples
N = 400      
# dimensionality of features
feats = 784  
#D[0] is random input dummy data (actual data), matrix with dimensions N x feats
#D[1] is random binary dummy data (expected output), vector with dimensions N 
D = (rng.randn(N, feats), rng.randint(size=N, low=0, high=2))
# declare Theano symbolic variables
x = T.matrix("x") #input
y = T.vector("y") #output
w =theano.shared(rng.randn(feats), name="w") #weights, intialize them with any random numbers
b = theano.shared(0., name="b") #bias, initialized with zero
# print w.get_value()
print D[0]
```

```python
#theano expression function
k = T.dot(x, w)-b
p_1 = 1 / (1 + T.exp(k))
prediction = p_1 > 0.5
# prediction = T.dot(x, y)
#cross-entropy loss function
xent= -y*T.log(p_1)-(1-y)*T.log(1-p_1) 
#cost function
cost = xent.mean() + 0.01 * (w**2).sum()
gw,gb= T.grad(cost, [w, b])
```

```python
#train our model:
# x = T.dmatrix('x')
learning_rate = 0.1
s = 1 / (1 + T.exp(-x))
train = theano.function(
          inputs=[x,y],
          outputs=[prediction, xent],
          updates=((w, w - learning_rate * gw), (b, b - learning_rate * gb)))
predict = theano.function(inputs = [x], outputs = prediction)
# Train
accuracy = []
epoch = 10
for i in range(epoch):
    print "epoch" + str(i)
    pred, err = train(D[0], D[1])
    print pred
    print D[1]
    
# print "Final model:"
# print pred
# print err
print "target values for D: ", D[1]
print "predictions on D: ", predict(D[0])
```
